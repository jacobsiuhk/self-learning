### Linear Regression

```python
data
```

```python
data['Transaction Date']=pd.to_datetime(data['Transaction Date'])
data['Transaction Date']=data['Transaction Date'].dt.strftime('%Y')
data['Transaction Date']=pd.to_numeric(data['Transaction Date'], downcast='signed')
```

```python
data.dtypes
```

```python
#one hot encoding, create dummy vairables for multi-categorical data
data.drop(['Estate'], axis=1, inplace=True)
#too many levels in estate, drop it before modelling

#drop unit price of GFA and SFA
data.drop(['Unit Price of Gross Floor Area'], axis=1, inplace=True)
data.drop(['Unit Price of Saleable Floor Area'], axis=1, inplace=True)
```

```python
# Create two data sets for numeric and non-numeric data
data_num = data.select_dtypes(exclude=['object'])
data_str = data.select_dtypes(include=['object'])
```

```python
from sklearn.preprocessing import StandardScaler
# define standard scaler
ss = StandardScaler()
data_num_stand = ss.fit_transform(data_num)
```

```python
data_num_stand = pd.DataFrame(data_num_stand,columns = ['Transaction Date','Transaction Price','Saleable Floor Area',
                                                        'Gross Floor Area','Floor','Inflation rate','Unemployment rate'])
```

```python
# One-hot encode the non-numeric columns
data_str_onehot = pd.get_dummies(data_str)

data_str_onehot.reset_index(drop=True,inplace=True)

# Union the one-hot encoded columns to the numeric ones
data_model = pd.concat([data_num_stand, data_str_onehot], axis=1)
```

```python
data_model
```

```python
# Create X data for the model
X = data_model.drop('Transaction Price', axis = 1)

# Create a set of y data for training
y = data_model[['Transaction Price']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=123)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
import numpy as np
coef = pd.DataFrame(np.transpose(reg.coef_), columns = ['Coefficient'])
```

```python
feature = pd.DataFrame(X.columns, columns = ['Feature'])
```

```python
combined=pd.concat([coef, feature],axis=1)
```

```python
combined
```

```python
combined.loc[combined['Coefficient']==np.max(combined['Coefficient'])]
#The standardized coefficient for district_kowloon station is the greatest. 
#So, it is the most important factor affecting property price.
```

```python
combined.loc[combined['Coefficient']==np.min(combined['Coefficient'])]
```

```python
y_pred = reg.predict(X_test)
```

```python
y_test=y_test.reset_index(drop=True)
```

```python
y_pred = pd.DataFrame(y_pred, columns = ['Predicted Price'])
```

```python
diff = pd.concat([y_test, y_pred],axis=1)
diff*data['Transaction Price'].std()+data['Transaction Price'].mean()
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
metrics.r2_score(y_test, y_pred)
#R^2 of the model
```
