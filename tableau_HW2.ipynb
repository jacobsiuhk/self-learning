# Data Preparation 2 - Removing useless columns

```python
pd.set_option('display.max_columns', None)
census_16_original
```

We notice that dc & dc_code are meaningless.<br>
Also, dc_eng and dc_chi are redundant, I would like retain the dc_eng column.

```python
unused_column=['dc','dc_code','dc_chi']
```

```python
census_16_original.drop(unused_column, axis=1,inplace=True)
```

# Data Preparation 3 - Converting columns of object type

After that, we need to explore which columns are of object type, so as to handle them in an approporiate way.

```python
census_16_original.loc[:, census_16_original.dtypes == object]
```

The reason for the remaining columns are of object is that, they contain missing values indicated by "-".

It should be reasonable to replace "-" by 0 by examining some subset of dataset.

For example, In out[14] of this notebook, the "-" in row 0 and column "oq_s" means there are no subsidised home ownership housing in central and western district, and this is not a strange case.


##### Removing "-":<br>

```python
census_16_original=census_16_original.replace(to_replace ="-", value ="0") 
```

We can convert the columns to numeric values after removing "-".

```python
census_16_original['oq_s']=pd.to_numeric(census_16_original['oq_s'])
census_16_original['dh_s']=pd.to_numeric(census_16_original['dh_s'])
census_16_original['pop_s']=pd.to_numeric(census_16_original['pop_s'])
census_16_original['dh_r0']=pd.to_numeric(census_16_original['dh_r0'])
```

```python
census_16_original.info()
```

The district name is the only column with object type now.

For the last row of the dataset, it represents the aggregation sum of a column.<br>
I chose to remove it to prevent aggregation error.

```python
census_16_original = census_16_original[:-1]
```

```python
census_16_original.to_csv('census_16_ready.csv',index=False)
```
